# Classification Problem

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
org_data = pd.read_csv('mushrooms.csv')
org_data
```

```python
org_data.columns
```

```python
from sklearn.preprocessing import LabelEncoder
en = LabelEncoder()
```

```python
# Label Encoding
data1 = org_data
for data in data1.columns :
    data1[data] = en.fit_transform(data1[data])  
```

```python
data1
```

```python
data1.isna().sum()
```

```python
X = data1.drop(['class'], axis=1)
y = data1['class']
```

```python
X
```

```python
y
```

```python
# applying feature engineering
from feature_engine.selection import DropDuplicateFeatures,DropCorrelatedFeatures,DropConstantFeatures,SmartCorrelatedSelection
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=10,n_jobs=4)
pip = Pipeline([('Constant',DropConstantFeatures(tol=0.98, variables=None)),('Duplicate',DropDuplicateFeatures(variables=None)),('Smart',SmartCorrelatedSelection(selection_method='variance',estimator=rf))])
xnew = pip.fit_transform(X)
xnew.shape
```

```python
# applying smote
from imblearn.over_sampling import SMOTE
smote_enn = SMOTE(random_state=42)
x_resampled,y_resampled = smote_enn.fit_resample(xnew,y)
x_resampled.shape,y_resampled.shape
```

# Applying 10 fold cross validation and Machine Learning Models 

```python
from numpy import mean
from numpy import std
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# prepare the cross-validation procedure
cv = KFold(n_splits=10, random_state=1, shuffle=True)
```

```python
# preparing train, test and spilt procedure
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
model1 = GaussianNB()
```

```python
# TTS Method
model1.fit(X_train,y_train)
```

```python
y_pred = model1.predict(X_test)
print("Accuracy:",accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n",confusion_matrix(y_test, y_pred))
print("Classification Report:\n",classification_report(y_test, y_pred))
```

```python
# K Fold Cross validation Method
# evaluate model
score1 = cross_val_score(model1, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(score1), std(score1)))
```

```python
model2 = KNeighborsClassifier()
```

```python
# TTS Method
model2.fit(X_train,y_train)
```

```python
y_pred_2 = model2.predict(X_test)
print("Accuracy:",accuracy_score(y_test, y_pred_2))
print("Confusion Matrix:\n",confusion_matrix(y_test, y_pred_2))
print("Classification Report:\n",classification_report(y_test, y_pred_2))
```

```python
# K Fold Cross validation Method
# evaluate model
score2 = cross_val_score(model2, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(score2), std(score2)))
```
